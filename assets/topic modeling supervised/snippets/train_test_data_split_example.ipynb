# Example train-test data split


Load the dataset

```python
import pandas as pd

path_to_dataset = 'pravda_first_semester_2024.tsv'
df = pd.read_csv(path_to_dataset, sep='\t')
df.head(2)
```

Get the features

```python
features = df['headline']
features
```

Create the Document-Term matrix

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

dtm = vectorizer.fit_transform(features)
dtm
```

Split the data into training and testing sets

```python
from sklearn.model_selection import train_test_split

X = dtm
y = df['section']
test_size = 0.3         # 30% of the data is reserved for testing purposes.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
```
